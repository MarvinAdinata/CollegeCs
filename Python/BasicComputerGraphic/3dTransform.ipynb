```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[2, 2, 10],
                   [3, 4, 8],
                   [4, 6, 12]])

translation_matrix = np.array([[1, 0, 0, 2],
                               [0, 1, 0, 2],
                               [0, 0, 1, 2],
                               [0, 0, 0, 1]])
# Rotate in x axis
angle = 180
radian = np.radians(angle)
rotation_matrix = np.array([[1, 0, 0, 0],
                            [0, np.cos(radian), -np.sin(radian), 0],
                            [0, np.sin(radian), np.cos(radian)  , 0],
                            [0, 0, 0, 1]])

scale_matrix = np.array([[2, 0, 0, 0],
                         [0, 2, 0, 0],
                         [0, 0, 2, 0],
                         [0, 0, 0, 1]])

translated_points = np.dot(translation_matrix, np.vstack((points.T, np.ones(len(points)))))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', label='Original Points')
ax.scatter(translated_points[0, :], translated_points[1, :], translated_points[2, :], c='r', label='Translated Points')
ax.plot3D(points[:, 0], points[:, 1], points[:, 2], c='b')
ax.plot3D(translated_points[0, :], translated_points[1, :], translated_points[2, :], c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([1, 1, 1])
ax.legend()
plt.show()

rotated_points = np.dot(rotation_matrix, np.vstack((points.T, np.ones(len(points)))))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', label='Original Points')
ax.scatter(rotated_points[0, :], rotated_points[1, :], rotated_points[2, :], c='r', label='Rotated Points')
ax.plot3D(points[:, 0], points[:, 1], points[:, 2], c='b')
ax.plot3D(rotated_points[0, :], rotated_points[1, :], rotated_points[2, :], c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([1, 1, 1])
ax.legend()
plt.show()

scaled_points = np.dot(scale_matrix, np.vstack((points.T, np.ones(len(points)))))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', label='Original Points')
ax.scatter(scaled_points[0, :], scaled_points[1, :], scaled_points[2, :], c='r', label='Scaled Points')
ax.plot3D(points[:, 0], points[:, 1], points[:, 2], c='b')
ax.plot3D(scaled_points[0, :], scaled_points[1, :], scaled_points[2, :], c='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_box_aspect([1, 1, 1])
ax.legend()
plt.show()
```

```python
import cairo

surface_size_x = 1280
surface_size_y = 720
pixel_width = 20

def drawGrid(ct, size_x, size_y, pixel_size):
    for i in range(int(size_x / pixel_size)):
        for j in range(int(size_y / pixel_size)):
            ct.rectangle(i*pixel_size, j*pixel_size, pixel_size, pixel_size)
    ct.stroke()

with cairo.SVGSurface("pixel_grid.svg", 600, 400) as surface:
    # Membuat Surface dan Context
    cairo.ImageSurface(cairo.FORMAT_RGB24, surface_size_x, surface_size_y)
    ctx = cairo.Context(surface)

    # Warna Background
    ctx.set_source_rgb(0.0, 0.0, 0.0)
    ctx.paint()
    
    # Warna Garis
    ctx.set_source_rgba(1.0, 1.0, 1.0)
    ctx.set_line_width(1)

    #Menggambar Pixel Grid
    drawGrid(ctx, surface_size_x, surface_size_y, pixel_width)

    surface.write_to_png('pixel_grid.png')

display(SVG(filename="pixel_grid.svg"))
```

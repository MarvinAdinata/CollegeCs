```python
import numpy as np
import pandas as pd
import matplotlib
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

def linearCongruentialMethod(seed, m, a, c, randomNums, noOfRandomNums):
    already_seen = [-1] * (noOfRandomNums)
    temp = ((seed * a) + c) % m
    randomNums[0] = round(temp / m, precision)
    for i in range(1, noOfRandomNums):
        temp = ((temp * a) + c) % m
        randomNums[i] = round(temp / m, precision)
        if randomNums[i] not in already_seen:
            already_seen[i] = randomNums[i]
        else:
            print("REPEAT AT = ", i)
            break
    pass

seed = 1
m = 2**16   # Modulus parameter
a = 75        # Multiplier term
c = 74        # Increment term
precision = 5
noOfRandomNums = 10**3
randomNums = [-1] * (noOfRandomNums)
linearCongruentialMethod(seed, m, a, c, randomNums, noOfRandomNums)
for i in range(0, noOfRandomNums, 1):
    print(randomNums[i])
    pass

data_lcg = pd.DataFrame(randomNums)
data_lcg.to_csv("data_lcg_1.csv")
pd.plotting.lag_plot(data_lcg)

plot_acf(data_lcg)
plt.show()
```
